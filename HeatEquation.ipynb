## 2 Derivation of the heat equation


The law of conservation of energy is the basis of the heat equation, which states that energy is not lost or created, but merely transformed from one form to another. Mathematically it looks like this:

The heat equation comes directly from the *law of conservation of energy*:  
energy cannot be created or destroyed, only moved from one place to another.

Consider a small piece of a rod in the interval $x, x+\Delta x$.  
The amount of thermal energy stored in this piece is proportional to the temperature: <br>


Consider a small piece of a rod in the interval $[x, x+\Delta x]$.
The amount of thermal energy stored in this piece is proportional to the temperature:
$
E = \rho c_p\, u(x,t)\, \Delta x .
$

Here,
- $\rho$ denotes the density,
- $c_p$ the specific heat capacity,
- $u(x,t)$ the temperature.


According to the conservation of energy, the rate of change of this energy equals the net heat flow into the interval.

Using Fourier’s law, the heat flow is  
$$
I = -k\,u_x,
$$
where $k$ is the thermal conductivity.

The net inflow of heat is  
$$
I(x) - I(x+\Delta x) \approx -I_x(x)\,\Delta x.
$$

Energy balance gives  
$$
\rho c_p\, u_t \,\Delta x = -I_x\, \Delta x.
$$

Canceling $\Delta x$ yields  
$$
\rho c_p\, u_t = -I_x.
$$

Using Fourier’s law again,  
$$
I_x = -(k\,u_x)_x,
$$
we obtain  
$$
u_t = \frac{k}{\rho c_p}\, u_{xx}.
$$

The constant  
$$
a^2 = \frac{k}{\rho c_p}
$$
is called the *thermal diffusivity*, and the equation becomes  
$$
u_t = a^2 u_{xx}.
$$



There are different notations, but all are equal. Here are some other notations:

$$
u_t(x,t) = a^2 u_{xx}(x,t)
$$

$$
u_t = a^2 u_{xx}
$$

$$
\frac{\partial u}{\partial t}=a^2\frac{\partial^2 u}{\partial x^2} 
$$

Solution for Thermal Diffusivity of Copper

Parameters:
\begin{align*}
    k &= 401 \, \text{W/(m$\cdot$K)} = 401 \, \text{J/(m$\cdot$K$\cdot$s)} \\
    \rho &= 8.940 \, \text{g/cm$^3$} = 8940 \, \text{kg/m$^3$} \\
    c_p &= 385 \, \text{J/(kg$\cdot$K)}
\end{align*}

Calculation of Thermal Diffusivity $a^2$
The thermal diffusivity $a^2$ for copper is calculated as follows:

\begin{align*}
    a^2 &= \frac{k}{\rho \cdot c_p} \\
        &= \frac{401}{8940 \times 385} \\
        &\approx 1.165 \times 10^{-4} \, \text{m$^2$/s}
\end{align*}

Converting to $mm^2/s$
\begin{align*}
    a^2 &\approx 1.165 \times 10^{-4} \, \text{m$^2$/s} \times 10^6 \, \text{mm$^2$/m$^2$} \\
        &\approx 116.5 \, \text{mm$^2$/s}
\end{align*}

```python
# Simple One sine mode (k=1): u(x,t) = sin(x) * exp(-a^2 t)
def u_sin_xt(x, t, a2):
    # Single sine mode with homogeneous Dirichlet Boundary conditions
    return np.sin(x) * np.exp(-a2 * t)  # Initial graph meets exponential decay time factor
```

### 3.4 1D heat equation (Fourier sine mode)

We already solved the PDE with Fourier in Task 4 and now it is time to simulate it.

```python
# Parameters of Fourier sinus modes:
a2 = 1.0        # thermal diffusivity a^2
N = 50          # number of Fourier terms
L = np.pi       # rod length (x in [0, pi])

nx = 300        # Number of measure points on x-axis
x_vals = np.linspace(0.0, L, nx)     # actual positions of the measure points from 0 to L # actual positions of the measure points from 0 to L

# Times for line plots and checks
plot_times = [0.0, 0.01, 0.1, 1.0]
```

<div class="alert alert-block alert-info">

### Task 7:  <font color=red>(5 points)</font>

Implement/Define the functions `coeffs()`, `u_xt()` and `u_grid()` with the solution from Task 4 for $C_k$. `coeffs()` has the target to calculate the Fourier sinus coefficient $C_k$, `u_xt()` has the target to calculate the temperature $u(x,t)$ at time $t$ and location $x$ based on the fourier series and `u_grid()` has the target to calculate $u(x,t)$ for all positions $x$ and time steps $t$ and stores the results in a 2D array.

<div class="alert alert-block alert-success">
    
<b>Solution of Task 7</b>


    
</div>

```python
# Calculates the fourier sinus coefficient c_k for f(x)=1
def coeffs(N):
    k = np.arange(1, N + 1, dtype=float)    # Create an array k with numbers from 1 to N (k=1 big waves, k=100 small waves)
    c_k = 2.0 * (1.0 - np.cos(k * np.pi)) / (np.pi * k)    # Calculate fourier coefficient c_k with given formula
    return k, c_k

# Calculates the temperature u(x,t) at time t and location x based on the fourier series 
def u_xt(x, t, a2, N):
    x_arr = np.asarray(x, dtype=float)  # Convert input x to array
    x_flat = x_arr.reshape(1, -1)   # Make an array with 1 row and as many columns as x_arr has elements
    k, c_k = coeffs(N)  # Get fourier coefficients
    k = k[:, None]  # Change array k to one column and as many rows as k has elements
    terms = c_k[:, None] * np.sin(k * x_flat) * np.exp(-a2 * (k ** 2) * t)  # Calculate the terms of the Fourier series for each k and x at the time t
    u_flat = terms.sum(axis=0)  # Add all the fourier terms together for each x to get the temperature u(x,t) at time t
    return u_flat.reshape(x_arr.shape)  # Reshape the flat array back to the original shape of x_arr (reshape important, when function is called with plt.plot)

# Calculates u(x,t) for all positions x and time steps t and stores the results in a 2D array
def u_grid(x_grid, t_grid, a2, N):
    U = np.empty((len(t_grid), len(x_grid)), dtype=float)   # Allocate empty 2D array for the results based on the length of t_grid and x_grid
    # Loop over all time steps with index i and time t from t_grid, and fill the row i and all columns with the temperature values 
    for i, t in enumerate(t_grid):
        U[i, :] = u_xt(x_grid, t, a2, N)
    return U
```

```python
# Checks if the numeric implementation satisfies boundary conditions and initial value approximation
for t in (0.01, 0.1, 1.0):
    left = u_xt(0.0, t, a2, N)
    right = u_xt(L, t, a2, N)
    print(f"t={t:.2f}: u(0,t)={left}, u(pi,t)={right}") # sin(k*pi) for example is not exactly zero, but extremely close to zero, which is acceptable for practical purposes

# Convergence check at t=0 for different N
plt.figure(figsize=(7.5, 4))
for N_test in (1, 10, 50, 200):
    plt.plot(x_vals, u_xt(x_vals, 0.0, a2, N_test), label=f"N={N_test}")
plt.plot(x_vals, np.ones_like(x_vals), "k--", label="u(x,0)=1")
plt.xlabel("x")
plt.ylabel("u(x,0)")
plt.title("Initial condition approximation")
plt.legend()
plt.show()

# Plot u(x,t) for every array value in plot_times
plt.figure(figsize=(7.5, 4))
for t in plot_times:
    plt.plot(x_vals, u_xt(x_vals, t, a2, N), label=f"t={t:g}")
plt.xlabel("x")
plt.ylabel("u(x,t)")
plt.title("Fourier sine mode solution at several times steps")
plt.legend()
plt.show()
```
